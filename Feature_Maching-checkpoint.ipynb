## Brute force detection with ORD Description

###  Brute force matching with ORD description
##### Brute force matcher is simple. it takes the descriptior of one feature in first set and is matches with all other feature in second set using some ditance calculation. and the closest one is returen.

#### What is BFMatcher() Function

###### What is BFMatcher() Function is used to 





```python
orb=cv2.ORB_create()

'''
orb= cv2.ORB_Create() create orb detector object.
orb.detectAndCompute(image,mask)is a method that detect key point annd computer description in given image.
it take the image (reese and cereals)
and an option mask( which is set to None here)
As an inputer take two op:kp1 which represent the keypoints and de1 which represent the description for the 1st image(resses).
'''

kp1,des1=orb.detectAndCompute(reeses,None)
kp2,des2=orb.detectAndCompute(cereals,None)

# create BFMatcher(cv2.NORM_HAMMING,crossCheck=True)
bf=cv2.BFMatcher(cv2.NORM_HAMMING,crossCheck=True)




'''
cv2.BFMatcher() is function of open cv library used for matching descriptionbetween two set of keypoint description.

cv2.BFmatcher: this function create a brute force matcher object in openCV. IT is simple but often effective method

Cv2.NORM_HAMMING: This is distance measured by used for comparing binary description
 crossCheck=True: Setting Crosscheck True that means matcher only returen match
'''


# mAtch descriptor
matches=bf.match(des1,des2)

# sort them in the order of ther  distance
matches=sorted(matches,key=lambda x:x.distance)


# draw 1st 25 lines
resses_matches=cv2.drawMatches(reeses,kp1,cereals,kp2,matches[:25],None,flags=2)










```

```python
dispaly(resses_matches)
```

# Brute force matchinge with SIFT descriptors and ratio test

```python
# create sift object
sift = cv2.xfeatures2d.SIFT_create()

# find the key point and description with SIFT
kp1,des1=sift.detectAndCompute(reeses,None)
kp2,des2=sift.detectAndCompute(cereals,None)

# BDMatcher with default params
bf=cv2.BFMatcher()
matches=bf.knnMatch(des1,des2,k=2)

# apply ratio test
good=[]
for match1,match2 in matches:
    if match1.distance < 0.75*match2.distance:
        good.append([match1])
        
# cv2.drawMatchesKnn expects list of list as matches
sift_matches= cv2.drawMatchesKnn(reeses,kp1,cereals,kp2,good,None,flags=2)
```

```python
dispaly(sift_matches)
```

## FLANN based matcher

#### we impliment feature based matching between two images using invarient features transform (SIFT) and (FLANN). The SIFT is used  fetures keypoint and descriptors.


```python
# create sift object
sift = cv2.xfeatures2d.SIFT_create()

# find the key point and description with SIFT
kp1,des1=sift.detectAndCompute(reeses,None)
kp2,des2=sift.detectAndCompute(cereals,None)

# FLANN parametr
FLANN_INDEX_KDTREE=0
index_params=dict(algorithm= FLANN_INDEX_KDTREE,trees=5)
search_params= dict(check=50)
flann=cv2.FlannBasedMatcher(index_params,search_params)

'''
FLANN_INDEX_KDTREE=0 : This line dfine algorithm is used to be is a KD-tree. Kd-Tree is data structure used to organizing points
in a k-dimensional space

indedx_params: this is dict containing parameter for the index.

'''





# BDMatcher with default params

matches=flann.knnMatch(des1,des2,k=2)

# apply ratio test
good=[]
for match1,match2 in matches:
    if match1.distance < 0.7*match2.distance:
        good.append([match1])
        
# cv2.drawMatchesKnn expects list of list as matches
flann_matches= cv2.drawMatchesKnn(reeses,kp1,cereals,kp2,good,None,flags=0)
dispaly(flann_matches)
```
